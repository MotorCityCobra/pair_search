```python
def get_sector_and_industry(ticker):
    stock = yf.Ticker(ticker)
    info = stock.info
    sector = info.get('sector', 'N/A')
    industry = info.get('industry', 'N/A')
    return sector, industry
```

```python
import yfinance as yf
import pandas as pd
from datetime import datetime
import numpy as np
from hmmlearn import hmm
import matplotlib.pyplot as plt
from scipy.ndimage import gaussian_filter1d
from statsmodels.tsa.stattools import adfuller, coint
from itertools import groupby
from operator import itemgetter
import warnings
import shutil
import os
plt.ioff()

# Initialize an empty DataFrame to store the data
SPY_df = pd.DataFrame()

today_date = datetime.now().strftime('%Y-%m-%d')
# Fetching data for each ticker and storing in the DataFrame
for ticker in sp:
    if ticker not in list(SPY_df.columns):
        print(ticker)
        data = yf.download(ticker, start='2022-01-01', end=today_date)  # Adjust start and end dates as needed
        scaled = (data['Adj Close'] - data['Adj Close'].min()) / (data['Adj Close'].max() - data['Adj Close'].min())

        SPY_df[ticker] = scaled  # Store Adjusted Adj Close prices

print(SPY_df)
```

```python
SPY_df.dropna(axis=1, how='all', inplace=True)
nan_percentage = SPY_df.isna().mean()
threshold = 0.05
SPY_df = SPY_df.loc[:, nan_percentage <= threshold]
SPY_df.fillna(method='ffill', inplace=True)
SPY_df.fillna(method='bfill', inplace=True)

print("Columns with NaNs:", SPY_df.columns[SPY_df.isna().any()].tolist())
print("Total NaNs remaining:", SPY_df.isna().sum().sum())
```

```python
sector_dict = {}
for ticker in SPY_df.columns:
    industry, _ = get_sector_and_industry(ticker)
    if industry == 'N/A':
        pass
    else:
        sector_dict[ticker] = industry

for x in SPY_df.columns:
    if x in sector_dict.keys():
        pass
    else:
        del SPY_df[x]
```

```python


d = {}
spread_df = pd.DataFrame()
for m in range(len(SPY_df.columns)):
    for r in range(len(SPY_df.columns[:m])):
        try:
#         if 1 == 1:
            if sector_dict[SPY_df.columns[m]] == sector_dict[SPY_df.columns[r]]:
#             if 1 == 1:
                print(SPY_df.columns[m], SPY_df.columns[r])
                spread_o = SPY_df[SPY_df.columns[m]] - SPY_df[SPY_df.columns[r]]

                spread_df[SPY_df.columns[m]+'-'+SPY_df.columns[r]] = spread_o

                scaled_data_0 = np.array(SPY_df[SPY_df.columns[m]])
                scaled_data_1 = np.array(SPY_df[SPY_df.columns[r]])

                # Calculate the spread
                spread = scaled_data_0 - scaled_data_1

                # Prepare data for HMM
                X = spread.reshape(-1, 1)

                # Create and fit HMM model
                model = hmm.GaussianHMM(n_components=3, covariance_type="full", n_iter=10000)
                model.fit(X)

                # Calculate state probabilities
                state_probabilities = model.predict_proba(X)
                smoothed_probabilities = np.apply_along_axis(lambda x: gaussian_filter1d(x, sigma=5), axis=0, arr=state_probabilities)
                transition_matrix = model.transmat_

                # Identify high-confidence regimes
                confidence_threshold = 0.8  # Adjust as needed
                high_confidence_mask = np.max(smoothed_probabilities, axis=1) > confidence_threshold

                # Find contiguous segments of high confidence
                segments = []
                for k, g in groupby(enumerate(high_confidence_mask), key=itemgetter(1)):
                    if k:
                        segment = list(map(itemgetter(0), g))
                        segments.append((segment[0], segment[-1]))

                # Define test functions
                def check_stationarity(series):
                    if len(series) <= 20:  # Adjust this threshold as needed
                        return False  # Consider short segments as non-stationary
                    if np.allclose(series, series[0]):  # Check if series is constant
                        return False  # Constant series is considered non-stationary
                    try:
                        with warnings.catch_warnings():
                            warnings.simplefilter("ignore")
                            result = adfuller(series, autolag='AIC')
                        return result[1] < 0.05  # Return True if p-value < 0.05 (stationary)
                    except (ValueError, Warning):
                        return False  # If test fails, consider it non-stationary

                def check_cointegration(series_1, series_2):
                    if len(series_1) <= 20 or len(series_2) <= 20:  # Adjust this threshold as needed
                        return False  # Consider short segments as non-cointegrated
                    if np.allclose(series_1, series_1[0]) or np.allclose(series_2, series_2[0]):
                        return False  # If either series is constant, they're not cointegrated
                    try:
                        with warnings.catch_warnings():
                            warnings.simplefilter("ignore")
                            _, p_value, _ = coint(series_1, series_2)
                        return p_value < 0.05  # Return True if p-value < 0.05 (cointegrated)
                    except (ValueError, Warning):
                        return False  # If test fails, consider it non-cointegrated

                # Perform stationarity and cointegration tests
                stationarity_results = []
                cointegration_results = []
                total_time = len(spread)
                stationarity_score = 0
                cointegration_score = 0

                for start, end in segments:
                    segment_data_0 = scaled_data_0[start:end]
                    segment_data_1 = scaled_data_1[start:end]
                    segment_length = end - start + 1

                    stationarity_result = check_stationarity(segment_data_0 - segment_data_1)
                    cointegration_result = check_cointegration(segment_data_0, segment_data_1)

                    stationarity_results.append(stationarity_result)
                    cointegration_results.append(cointegration_result)

                    if stationarity_result:
                        stationarity_score += segment_length / total_time
                    if cointegration_result:
                        cointegration_score += segment_length / total_time

                total_score = stationarity_score + cointegration_score
                d[SPY_df.columns[m]+'-'+SPY_df.columns[r]] = total_score

                # Plotting
                fig = plt.figure(figsize=(15, 12), num=1)  # Use figure number 1
                # Original series and spread plot
                plt.subplot(3, 1, 1)
                plt.plot(scaled_data_0, label='Series 0', alpha=0.7)
                plt.plot(scaled_data_1, label='Series 1', alpha=0.7)
                plt.plot(spread, label='Spread', color='gray', alpha=0.5)
                plt.title(SPY_df.columns[m]+'-'+SPY_df.columns[r])
                plt.legend()

                # Highlight high-confidence segments and add test results
                for (start, end), stationarity, cointegration in zip(segments, stationarity_results, cointegration_results):
                    plt.axvspan(start, end, alpha=0.2, color='yellow')
                    # Plot stationarity result (circle)
                    plt.plot(start, spread[start], 'go' if stationarity else 'ro', markersize=10)
                    # Plot cointegration result (triangle) with vertical offset
                    triangle_y_position = spread[start] + 0.1  # Adjust as needed
                    plt.plot(start, triangle_y_position, 'g^' if cointegration else 'r^', markersize=10)

                # State probabilities plot
                plt.subplot(3, 1, 2)
                for i in range(model.n_components):
                    plt.plot(smoothed_probabilities[:, i], label=f'State {i} Probability')
                plt.xlabel('Time')
                plt.ylabel('Smoothed State Probability')
                plt.title('Smoothed Probability of Being in Each State Over Time')
                plt.legend()

                # Transition matrix and test results legend
                plt.subplot(3, 1, 3)
                plt.axis('off')
                trans_text = "Transition Matrix:\n" + np.array2string(transition_matrix, precision=2, separator=', ')
                plt.text(0.05, 0.6, trans_text, fontsize=8, va="top")
                plt.text(0.05, 0.3, "● - Stationarity Test (Green: Pass, Red: Fail)", fontsize=10)
                plt.text(0.05, 0.2, "▲ - Cointegration Test (Green: Pass, Red: Fail)", fontsize=10)

                # Add score text to the plot
                score_text = f"Stationarity Score: {stationarity_score:.2f}\n"
                score_text += f"Cointegration Score: {cointegration_score:.2f}\n"
                score_text += f"Total Score: {total_score:.2f}"
                plt.text(0.05, 0.1, score_text, fontsize=14, transform=plt.gcf().transFigure)
                plt.text(0.2, 0.1, sector_dict[SPY_df.columns[m]], fontsize=14, transform=plt.gcf().transFigure)
                # plt.tight_layout()
                # plt.show()
                
                if segments and stationarity_results and cointegration_results:
                    last_segment = segments[-1]
                    last_stationarity = stationarity_results[-1]
                    last_cointegration = cointegration_results[-1]

                    if last_stationarity and last_cointegration:
                        # Add notification to the chart
                        plt.text(0.95, 0.05, "Recent: Stationary & Cointegrated", 
                                 fontsize=12, color='green', ha='right', va='bottom', 
                                 transform=plt.gca().transAxes)

                        # Save an extra copy in a new folder
                        extra_folder = 'E:/spread_charts/basket_570/recent_stationary_cointegrated/'
                        os.makedirs(extra_folder, exist_ok=True)
                        plt.savefig(extra_folder + SPY_df.columns[m] + '-' + SPY_df.columns[r] + '.png')
#                         plt.close(fig)

                plt.savefig('E:/spread_charts/basket_570/'+SPY_df.columns[m]+'-'+SPY_df.columns[r]+'.png')
                plt.close('all')

                # Print the total score
                print(f"Total Score (Stationarity + Cointegration): {total_score:.2f}")
        except:
            pass
```
